```python
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_addons as tfa
import math as m
import os
import datetime
import pickle
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import metrics


gpu = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpu[0], True)
```

```python
CATEGORIES = ['Black-grass', 'Charlock', 'Cleavers', 'Common Chickweed', 'Common wheat', 'Fat Hen', 'Loose Silky-bent',
              'Maize', 'Scentless Mayweed', 'Shepherds Purse', 'Small-flowered Cranesbill', 'Sugar beet']

dict_cat = {}

for i in range(len(CATEGORIES)):
    dict_cat[CATEGORIES[i]] = i

print(dict_cat)

              
NUM_CATEGORIES = len(CATEGORIES)
```

```python
indices = tf.range(len(CATEGORIES)-1, dtype = tf.int64)
table_init = tf.lookup.KeyValueTensorInitializer(CATEGORIES[0:11], indices-1)
num_oov_buckets = 1
table = tf.lookup.StaticVocabularyTable(table_init, num_oov_buckets)
```

```python

Category_count = []

for category in CATEGORIES:
    category_ds_filename = tf.data.Dataset.list_files('plant-seedlings-classification/train/' + category + '/*.png')
    c = [element.numpy() for element in category_ds_filename]
    Category_count.append(len(c))
    print( category, ' ',len(c))


_, ax = plt.subplots(1,figsize=(20, 5), facecolor='Silver')
for i in range(1):  
    ax.bar(CATEGORIES, Category_count )
    ax.set_xticklabels(CATEGORIES, rotation = 'vertical')
    ax.set_title("Class Vs Count")
```
